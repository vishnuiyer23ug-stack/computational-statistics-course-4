### Functions and If Statements

```python
def has_student_passed(scores): # After every condition, we put a :
  """
  This is going to be a multi line comment.
  This function takes the score and returns pass if the score
  is above forty.
  """
  if scores >= 40:
    return "The student has passed" # Executed if if condition evaluated true
    """
    return does not necessarily need to have brackets
    """
  else:
    return("The student has failed") # Executed if if condition evaluated false
Pass = [has_student_passed(i) for i in scores] # Creates a list that displays the condition for each score in our scores
Pass
```

### For loop

```python
for i in scores: # You CANNOT put a bracket after for
  print(f"i = {i}")
```

```python
for i in scores:
  print(f"i = {i}, {has_student_passed(i)}")
```

### Homework -Write a Function to calculate the Grade of the Student and return the Grade

```python
# Use either if and elseif, or look up the match function
def student_grades(scores):
  if scores <= 40:
    return "U"
  elif scores >40 and scores <=50:
    return "E"
  elif scores >50 and scores <=60:
    return "D"
  elif scores >60 and scores <=70:
    return "C"
  elif scores >70 and scores <=80:
    return "B"
  elif scores >80 and scores <=90:
    return "A"
  elif scores >90 and scores <=100:
    return "O"

for i in scores: # Using the function to print the scores
  print(f"score = {i}, {student_grades(i)}")
```

### Aggregate Functions

```python
print(scores_arr.mean()) # Calculates the mean
print(np.mean(scores_arr)) # Also calculates the mean. If we used np.mean on a list it would first convert to an array and then carry out the operation
print(np.min(scores_arr))
print(np.max(scores_arr))
print(np.argmin(scores_arr)) # Gives us the index of the minimum value
```

```python
scores_arr > 50 # This returns boolean values in an array itself, depending on the satisfaction of the condition
```

```python
scores_arr[scores_arr > 50] # Returns a new list with satisfaction of the conditions. The actual scores_arr remains the same with all the elements
```

### Slicing Arrays and Accessing Elements

```python
food_prices = np.array([530,293,200,501,293,706,77,291])
print(food_prices[1]) # Prints the second element
print(food_prices[3:6]) # Prints the fourth to six elements. The seventh element (index 6) is excluded, but the start index is included
print(food_prices[3:]) # Prints the fourth element to the end of the array
print(food_prices[:5]) # Prints everything until the sixth element
print(food_prices[:-1]) # This prints everything until the last element. -1 denotes the last element
print(food_prices[-1:-3]) # This is not a viable way to reverse the elements of a list
print(food_prices[::2]) # This has no start and end, but the number 2 indicates the step that you have to jump. So, this prints every second element
print(food_prices[::-1]) # This actually prints the list in reverse because the step is -1
```

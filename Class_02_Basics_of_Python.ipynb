### Exercise - Calculating the average score of the class

### Functions and If Statements

```python
def has_student_passed(scores): # After every condition, we put a :
  """
  This is going to be a multi line comment.
  This function takes the score and returns pass if the score
  is above forty.
  """
  if scores >= 40:
    return "The student has passed" # Executed if if condition evaluated true
    """
    return does not necessarily need to have brackets
    """
  else:
    return("The student has failed") # Executed if if condition evaluated false
Pass = [has_student_passed(i) for i in scores] # Creates a list that displays the condition for each score in our scores
Pass
```

### For loop

```python
for i in scores: # You CANNOT put a bracket after for
  print(f"i = {i}")
```

```python
for i in scores:
  print(f"i = {i}, {has_student_passed(i)}")
```

### Homework -Write a Function to calculate the Grade of the Student and return the Grade

```python
# Use either if and elseif, or look up the match function
def student_grades(scores):
  if scores <= 40:
    return "U"
  elif scores >40 and scores <=50:
    return "E"
  elif scores >50 and scores <=60:
    return "D"
  elif scores >60 and scores <=70:
    return "C"
  elif scores >70 and scores <=80:
    return "B"
  elif scores >80 and scores <=90:
    return "A"
  elif scores >90 and scores <=100:
    return "O"

for i in scores: # Using the function to print the scores
  print(f"score = {i}, {student_grades(i)}")
```
